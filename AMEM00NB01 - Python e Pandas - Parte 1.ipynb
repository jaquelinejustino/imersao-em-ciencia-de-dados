### 2.2. O que é Jupyter?
"Jupyter" é um acrônimo que significa Julia, Python, and R. Essas linguagens de programação foram as primeiras utilizadas na aplicação Jupyter, mas, atualmente, o notebook também suporta  [muitas outras linguagens](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels).

Por se tratar de uma aplicação server-client, o Jupyter Notebook App permite ao usuário fazer a edição e execução de seus notebooks via web browser (Chrome, por exemplo). A aplicação pode ser executada em um PC sem acesso a Internet. 

- É possível executar um documento notebook step-by-step (uma célula por vez) pressionando shift + enter.
- É possível executar todo o notebook em um único passo. Para isso, basta clicar no menu Cell -> Run All.
- Para reiniciar o kernel (i.e., o engine), clique no menu Kernel -> Restart. Isso pode ser útil quando você quer reiniciar o processamento desde o começo (nesse caso, as variáveis são deletadas, arquivos abertos são fechados, etc...).

**Keyboard Navigation**

Essa interface do Jupyter Notebook foi **otimizada para uso eficiente do teclado**. Isso é possível porque há dois modos diferentes de atalhos com o teclado: modo de edição (edition mode) e modo de comando (command mode).

Os atalhos mais importantes são **"Enter"**, que entra no **modo de edição**, e **"Esc"**, que entra no **modo de comando**.

No modo de edição, a maior parte do teclado está disponível para digitar nas células do notebook. Por conta disso, neste modo não há muitos atalhos disponíveis. Para sair do modo de edição, você pode entrar no modo de comando (pressionando ```Esc```), ou pode executar a célula (pressionando ```Ctrl+Enter```). Já no modo de comando, o teclado inteiro pode ser utilizado para atalhos, portanto, existem muito mais atalhos. Se você clicar em *Help->Keyboard Shortcuts*, será possível ver toda a lista de atalhos disponível.

Recomenda-se aprender os principais atalhos do modo de comando na seguinte ordem:

Navegação básica:
- ```Ctrl+Enter```: executa a célula
- ```Shift+Enter```: executa a célula e seleciona a próxima célula
- ```Up```/```k```: seleciona célula acima
- ```Down```/```j```: seleciona célula abaixo

Ações:
- ```s```: salva o Notebook
- ```y```, ```m```, ```1-6```: alterna o tipo da célula (code, markdown, títulos/subtítulos, respectivamente)
- ```a```: cria uma célula acima da selecionada
- ```b```: cria uma célula abaixo da selecionada
- ```x```: recorta a célula
- ```c```: copia a célula
- ```v```: cola a célula abaixo da atual
- ```d``` (pressionada rapidamente duas vezes): apaga a célula selecionada
- ```z```: desfaz a ação de apagar a célula
- ```l```: ativa a numeração das linhas

Operações no kernel:
- ```i``` (pressionada rapidamente duas vezes): interrompe o kernel
- ```0``` (pressionada rapidamente duas vezes): reinicia o kernel

Apesar de esses atalhos representarem o que você mais utilizará no cotidiano, existem algumas [dicas e macetes](https://www.dataquest.io/blog/jupyter-notebook-tips-tricks-shortcuts) que podem melhorar sua experiência com a aplicação Jupyter Notebook.

### 2.3. Como o Python funciona?

Antes de começarmos a aprender a linguagem Python, é importante entender alguns conceitos comumente negligenciados de como o Python interpreta o texto (código) que nós escrevemos em um editor e o transforma em um programa executável. A imagem a seguir resume esses conceitos:

![Python Interpreter](https://i.imgur.com/PJME67T.png)

Após isso, o **compilador** traduz o código fonte em 0's e 1's, que nós chamamos de **byte-code**. A **Máquina Virtual (VM)** então executa as instruções contidas no [byte-code](https://pt.wikipedia.org/wiki/Bytecode), que gera um programa executável. #### 2.6.2 Operações lógicas (booleana)
Python também suporta operadores lógicos e eles são realmente úteis quando estamos lidando com testes condicionais no código (não se preocupe, iremos explicar esse tópico mais pra frente). Essas operações lógicas retornam apenas resultados lógicos (booleanos) e elas podem ser classificadas nas seguintes categorias:

**Comparações**

Como o próprio nome sugere, essas operações comparam dois objetos e são muito úteis para determinar o comportamento do seu código em vários cenários (são muito úteis para executar o código bloco a bloco e entender se ele está fazendo o que é esperado).

````python
#Igual [==]. O operador == compara os VALORES dos operandos e checa a igualdade de VALORES
print(10 == 10)
print("Hi" == "Hello")
print (10.0 == 10) #Por que essa operação retorna verdadeiro? Porque o operador == compara VALORES e não TIPOS
````

````python
#Maior [>] (ou igual [>=]) que
print(5 > 3)
print(5.0 > 5) 
````

````python
#Menor [<] (ou igual [<=]) que
print(5 < 3)
print("bcdefghyjklmnopqrstuvwxyz" < "a") #Por que essa operação retorna falso? Porque o Python compara strings char por char! 
````

````python
#Diferente de [!=]
print("Hi" != "Bye")
print ("CASE SENSITIVE" != "Case Sensitive")
````

**Bitwise**

Basicamente essas operações funcionam similarmente aos clássicos AND, OR, NOT, etc. Elas são chamadas de bitwise porque executam a operação bit a bit!

````python
# AND (&)
print(False & False)
print(True & True)
print(True & True & True & True & False & True & True & True  & True)
````

````python
# OR (|)
print(True | False)
print(False | False)
print(False | False & False | True | False | False | False | False | False)
````

````python
# NOT (~)
print(bool(~False)) #bool() converte o tipo de um dado objeto em booleano. Nós iremos falar sobre isso mais tarde!
````

````python
#XOR (^). (eXclusive OR)
print(False ^ (False ^ True))
print(False ^ False ^ True)
print(False ^ True ^ False ^ False ^ False)
print(True ^ True ^ False) 
print(False ^ True ^ False & True)
print(False ^ True ^ False | True)

# Dica: pesquise sobre a precedência dos operadores quando estiver lidando com vários operadores para saber a ordem que o Python resolve as expressões
````
