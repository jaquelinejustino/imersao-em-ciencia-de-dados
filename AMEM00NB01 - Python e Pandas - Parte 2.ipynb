### 2.7. Variáveis e Tipos

Em Python, uma variável nada mais é que um **label**, um nome. A variável é criada no momento que você atribui um valor a ela. **Assignment (atribuição de valor)** é feita com sinal de igual simples (=):

````python
# Criando uma variável x ao atribuir um valor a ela

x = 5 # Lê-se como algo do tipo “a x é atribuído o valor 5".

# Uma vez que a atribuição é feita, a variável poderá ser usada em qualquer expressão, e seu valor poderá ser substituído
print(x)
````

#### 2.7.1. Nomes de variáveis
Uma variável pode ter um nome curto (como x ou y) ou um nome mais descritivo (idade, nome_carro, volume_total). [Regras para nomes de variáveis](https://www.python.org/dev/peps/pep-0008/#function-and-variable-names) no Python podem ser sintetizadas da seguinte forma:
- Um nome de variável deve começar com uma letra ou o caractere underscore

- Um nome de variável deve conter apenas caracteres alfanuméricos e underscores (A-z, 0-9, and _ )
- Nomes de variáveis são case-sensitives (age, Age e AGE são três variáveis distintas)

````python
# Criando três variáveis utilizando as convenções de nomes
name = "Jose da Silva"
age  = 27
likes_python = True

# Printando as variáveis
print ("Nome: ", name)
print ("Idade:  ", age)
print ("Gosta de Python? ", likes_python)
````

#### 2.7.3. Outros Tipos de Variáveis
Nas seções anteriores, nós lidamos com números inteiros, números não-inteiros e strings (textos). Esses são alguns de muitos [tipos de dados do Python](https://docs.python.org/3/library/datatypes.html), mas existem quatro principais tipos de dados:

- **int**: números inteiros. Ex: 0, 100, -3, etc;
- **float**: um número que possui tanto uma parte inteira quanto uma parte fracionária, separada por uma casa decimal (representada por um ponto .). Ex: 0.5, 0.33, -0.1, etc;
- **str**: também chamada de string, é um tipo que representa texto. Você pode utilizar tanto aspas simples quantos aspas duplas para construir uma string. Ex: "Hi", 'Hello'
- **bool**: representa valores lógicos e só pode assumir dois valores, *True* ou *False*.

````python
# Integer (int)
integer_number = 1

# Floating point (float)
float_number = 1.0

# String (str)
text = "Hello"

# Boolean (bool)
python_is_fun = True
````

Para descobrir o tipo de um valor ou da variável que se refere àquele valor, você pode usar a função type()!

````python
# Variable types
print(type(integer_number))
print(type(float_number))
print(type(text))
print(type(python_is_fun))
````

Mas como o Python assimila o tipo de uma variável? Bem, ele infere a partir do valor atribuído àquela variável. Essas linguagens de programação são conhecidas como [loosely typed programming languages](https://en.wikipedia.org/wiki/Strong_and_weak_typing).

### 2.8. Estruturas de decisão
Até agora, todos os exemplos seguiram um fluxo contínuo, onde todas as linhas de código de uma mesma célula foram executadas. Entretanto, em aplicações reais, isso raramente é o que acontece e, nesses casos, nós temos que usar o que chamamos de **estruturas de decisão** para redirecionar o fluxo de execução baseado em critérios condicionais que guiam o programa. Para fazer isso no Python, nós usamos **if**, **else** e **elif** statements, que possuem a seguinte estrutura genérica:

````python
# if(condição lógica 1 ):
#    executa esse bloco de código se a condição lógica 1 for verdadeira
#    bloco de código
#    bloco de código
# elif(condição lógica 2):
#    executa esse bloco de código se a condição lógica 2 for verdadeira
#    bloco de código
#    bloco de código
# elif(condição lógica 3):
#    executa esse bloco de código se a condição lógica 3 for verdadeira
#    bloco de código
#    bloco de código
#         .
#         .
#         .
# else:
#    executa esse bloco de código se nenhuma das condições lógicas anteriores forem verdadeiras
#    bloco de código
#    bloco de código
````

Antes que continuemos, é importante observar que, no Python, não existe nenhum caractere que indique um bloco de código, como há em muitas outras linguagens de programação (comumente o ;). Na verdade, o que o Python considera como o início de um novo bloco de código está baseado apenas na **indentação/tabulação** do código!!! **Isso é muito importante para o Python! Preste sempre atenção na indentação dos seus scripts, pois um tab a mais (ou a menos) pode mudar toda a lógica do seu código!** No mundo dos negócios, não queremos que um erro causado por isso gere um resultado catrastrófico, não é mesmo?

Observe a seguinte imagem e exemplo abaixo:
![Identation](https://i.imgur.com/qQ6zBHk.png)

````python
# Programa que diz se um aluno vai passar na disciplina, dada a sua nota final
grade = 6 # Nota final do aluno

# Estrutura de decisão/lógica
if (grade >= 7):
    print("Aprovado!")
elif ((grade < 7) & (grade >= 5)):
    print("Recuperação")
else:
    print("Reprovado!")
````

### 2.9. Estruturas de laço (Looping structures)

Na seção anterior, introduzimos o conceito em que você pode determinar quais blocos de código serão executados baseado em critérios condicionais. Entretanto, em muitas ocasiões, queremos continuar fazendo esses testes condicionais para ver se o sistema está mudando com o tempo. Existem, então, dois tipos de estruturas de laço: o ```while``` e o ```for```.

#### 2.9.1. **While**

O comando while continua em loop enquanto a condição de teste continua retornando True.

````python
# Variável contadora
counter = 1

# While loop
while counter <= 3: # Testa se counter <= 3. Se True, executa o bloco de código. Caso contrário, sai do loop
    print(counter)
    counter = counter + 1
````

#### 2.9.2. For
Esse comando executa o loop por um determinado número de vezes. Normalmente, utiliza-se a função **range** para determinar quantos loops serão executados.

````python
# range(start, stop)
selected_range = range(1,4)
print("Selected range = ", *selected_range) # O operador * "desempacota" a sequência previamente definida

# For loop
for counter in selected_range:
    print(counter)
````

#### 2.9.3. Interruptions
É possível interromper a execução de um loop com os seguintes comandos:
- **break**: cancela o loop
- **continue** : recomeça o loop a partir do próximo contador

````python
# For loop
for counter in range(1,11):
    if counter == 7:
        break; # Para o looping quando counter == 7
    print(counter)
````

````python
# For loop
for counter in range(1,11):
    if counter == 7:
        continue; # Pula o loop quando counter == 7
    print(counter)
````

#### 2.9.4. Try Except

Até o momento, as mensagens de erro foram apenas mencionadas e você provavelmente recebeu algumas ao tentar fazer os exercícios propostos. Agora que você já sabe sobre as estruturas de laço, podemos aprender a lidar com as mensagens de erros, que podem ser basicamente de dois tipos: *syntax erros* e *exceptions*.

Os **syntax erros** (erros de sintaxe) também são conhecidos como *parsing errors* e correspondem aos erros mais comuns que você vai receber enquanto aprende Python. Observe um exemplo abaixo.

````python
# Print com erro de sintaxe
while counter <= 3
    print(counter)
    counter = counter + 1
````

Observe que a linha com o erro é indicada, e há uma pequena seta apontando ao ponto em que o erro foi detectado, de forma que podemos rapidamente identificar onde está o problema e resolvê-lo. Neste exemplo, o erro acontece pois está faltando o ```:``` antes do bloco de código a ser executado.

Mesmo que a sintaxe do seu código esteja correta, pode ocorrer um erro quando você tenta executá-lo. Esses erros durante a execução são chamados de **exceptions** e vamos te mostrar uma forma de lidar com eles no Python. Observe os exemplos de *exception* abaixo.

````python
print(10 * (1/0))
````

````python
print(4 + numero_armazenado*5)
````

````python
print('5' + 7)
````

Observe que a última linha do erro indica o que aconteceu. As *exceptions* podem ser de diferentes tipos, o qual é printado como parte da mensagem. Nestes exemplos, os tipos são: *ZeroDivisionError*, *NameError* e *TypeError*.

Para lidar com as exceções, podemos utilizar o statement *try* dentro de um loop. O exemplo abaixo pede que o usuário insira um número, e realiza um loop até que o número inserido seja um inteiro.

````python
while True:
    try:
        x = int(input("Please enter a number: "))
        break
    except ValueError:
        print("Oops!  That was no valid number.  Try again...")
````

No exemplo citado, o statement *try* funciona da seguinte maneira: primeiro, o bloco de código dentro do *try* é executado (solicitando um número), e, caso não ocorra uma *exception*, seu bloco é pulado e a execução do *try* é finalizada (e o loop do while é interrompido pelo *break*, cujo funcionamento será explicado na seção seguinte). 

Caso uma *exception* ocorra (com a inserção de um caractere não numérico), o resto do bloco do *try* é ignorado e o bloco do *except* é executado (mostrando a mensagem de que não foi inserido um número válido).

Mas **atenção**! Tome muito cuidado ao utilizar o try except, pois você pode criar um **loop infinito**, que só é interrompido ao reiniciar o kernel do notebook (perdendo todas as variáveis criadas e dados carregados). Isso pode ocorrer pois a interrupção do kernel durante a execução de uma célula gera uma *exception*, e caso não especifiquemos a *exception* a ser tratada, podemos fazer com que o *try* seja executado repetidamente. Observe abaixo a alteração realizada no exemplo anterior:

while True: <br>
&nbsp;&nbsp;&nbsp;&nbsp;try: <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;x = int(input("Please enter a number: ")) <br>
&nbsp;&nbsp;&nbsp;&nbsp;except: <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print("Oops!  That was no valid number.  Try again...") <br>

Se executássemos uma célula com esse código, por mais que digitássemos um número válido, o loop do while continuaria (visto que não há uma interrupção, como um *break*). Além disso, se digitássemos um caractere não numérico ou tentássemos interromper a execução da célula, teríamos uma *exception*, porém a ocorrência da mesma executa apenas um print informando para o usuário tentar novamente e em seguida retorna à execução do *try*. Dessa forma, o loop só pode ser interrompido se o kernel for reiniciado. Por isso, sempre utilize um *break* no seu loop ou especifique a *exception* a ser tratada.
